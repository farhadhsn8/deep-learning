```python
import cv2
import matplotlib.pyplot as plt
```

```python

n_clusters = 8
```

```python
img = cv2.imread("input.jpg")

img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
plt.imshow(img)

h = img.shape[0]
w = img.shape[1]
```

```python
img = img.reshape((-1,3))
img.shape

```

```python
from sklearn.cluster import KMeans
import numpy as np
X = img
kmeans = KMeans(n_clusters = n_clusters, random_state=8888 , verbose=0).fit(X)
labels = kmeans.labels_
```

```python
centers = kmeans.cluster_centers_.astype(np.int16)
centers

img.shape
```

```python
img[1521]
```

```python
for i in range(img.shape[0]):
    img[i] = centers[labels[i]]
```

```python
img.shape
```

```python
img = img.reshape((h,w,3))

plt.imshow(img)
```
